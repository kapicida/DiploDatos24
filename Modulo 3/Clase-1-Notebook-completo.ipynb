```python
import matplotlib.pyplot as plt
import numpy as np
```

Vamos a definir un modelo sencillo: y = a*x

```python
x = [0.5, 2.3, 2.9]
y = [1.4, 1.9, 3.2]
a = 0
learning_rate = 0.1
```

```python
def gradiente(x,y):
    suma = 0
    n = len(x)
    for i in range(n):
        suma += a*x[i]**2-x[i]*y[i]
    return (1/n)*suma
```

```python
print('El valor de a es: ', a)
pendientes = []

for i in range(20):
    grad = gradiente(x,y)
    step = grad*learning_rate
    a = a- step
    print('El nuevo valor de a es: ',a,'      El step es: ', step)
    pendientes.append(a)
```

```python
# Definir el rango de valores para el eje x
x1 = np.linspace(0, 10, 400)

# Definir el mapa de colores
cmap = plt.get_cmap('viridis')

# Generar una secuencia de colores
num_colores = len(pendientes)
colores = [cmap(i / num_colores) for i in range(num_colores)]

# Graficar cada recta con su pendiente y color correspondiente
for pendiente, color in zip(pendientes, colores):
    y1 = pendiente * x1
    plt.plot(x1, y1, color=color)

# Añadir el scatter plot
plt.scatter(x, y, color='red', label='Datos originales')

# Añadir el colorbar
norm = plt.Normalize(vmin=0, vmax=num_colores)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Hay que definirlo para que funcione
cbar = plt.colorbar(sm, ax=plt.gca())
cbar.set_label('Epoch')

# Añadir título y etiquetas
plt.title('Adaptación de la pendiente a los datos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Añadir una leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

# Descenso del gradiente para dos parámetros
Ahora nuestro modelo es: y = ax +b

```python
x = [0.5, 2.3, 2.9]
y = [1.4, 1.9, 3.2]
n = len(x)

#Inicializamos los parámetros
a= 0
b= 0

learning_rate = 0.1
```

```python
def gradiente(x,y):
    da = 0
    db = 0
    for i in range(n):
        da += a*x[i]**2 -x[i]*y[i] + b*x[i]
        db += a*x[i] +b -y[i]
    da *= (1/n)
    db *= (1/n)
    return da, db
```

```python
print('El valor de a es: ', a, '         El valor de b es: ', b)
pendientes = []
ordenadas = []

for i in range(200):
    grad = gradiente(x,y)
    step_a = grad[0]*learning_rate
    step_b = grad[1]*learning_rate
    a = a -step_a
    b = b -step_b
    print(f'El nuevo valor de a es: {a}     step size: {step_a}')
    print(f'El nuevo valor de b es : {b}     step size = {step_b}')
    pendientes.append(a)
    ordenadas.append(b)
```

```python
#Calculo de los parámetros según Cuadrados Mínimos

x_prom = np.mean(x)
y_prom = np.mean(y)
m1 =0
m2 = 0
for i in range(n):
    m1 += (x[i]-x_prom)*(y[i]-y_prom)
    m2 += (x[i]-x_prom)**2
m = m1/m2
o = y_prom -m*x_prom

print(m, o)
```

```python
# Definir el rango de valores para el eje x
x1 = np.linspace(0, 10, 400)

# Definir el mapa de colores
cmap = plt.get_cmap('viridis')

# Generar una secuencia de colores
num_colores = len(pendientes)
colores = [cmap(i / num_colores) for i in range(num_colores)]

# Graficar cada recta con su pendiente y color correspondiente
for pendiente, ordenada, color in zip(pendientes, ordenadas, colores):
    y1 = pendiente * x1 + ordenada
    plt.plot(x1, y1, color=color)

# Añadir el scatter plot
plt.scatter(x, y, color='red', label='Datos originales')

# Add a colorbar
norm = plt.Normalize(vmin=0, vmax=num_colores)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Hay que definirlo para que funcione
cbar = plt.colorbar(sm, ax=plt.gca())
cbar.set_label('Epoch')

# Añadir título y etiquetas
plt.title('Adaptación de la recta a los datos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Añadir una leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
rectax = np.ones(len(pendientes))
rectax *= m
rectab = np.ones(len(ordenadas))
rectab *= o

fig, axs = plt.subplots(2,1, figsize=(6,10))
axs[0].plot(pendientes)
axs[0].plot(rectax)
axs[0].set_title('Optimización de la pendiente')
axs[0].set_xlabel('Iteraciones')
axs[0].set_ylabel('Valor de la pendiente')

axs[1].plot(ordenadas)
axs[1].plot(rectab)
axs[1].set_title('Optimización de la ordenada al orígen')
axs[1].set_xlabel('Iteraciones')
axs[1].set_ylabel('Valor de la ordenada al orígen')
fig.tight_layout()
```

## Optimizacion de una recta con valores ruidosos

```python
def gradiente(x,y):
    da = 0
    db = 0
    n = len(x)
    for i in range(n):
        da += a*x[i]**2 -x[i]*y[i] +b*x[i]
        db += a*x[i] +b -y[i]
    da *= (1/n)
    db *= (1/n)
    return da,db

def error(x,y):
    suma = 0
    n= len(x)
    for i in range(n):
        suma += (y[i]-a*x[i]-b)**2
    return (1/(2*n))*suma
```

```python
x = np.linspace(0,10,100)
a0 = -2
b0 = 10
y = a0*x + b0
y += np.random.randn(len(y))

plt.scatter(x,y)
plt.show()
```

```python
#Inicializamos los parámetros
param = np.random.normal(loc=0.0, scale=0.1, size = 2)
#loc = media,      scale= desv estandars
a = param[0]
b = param[1]

learning_rate = 0.01

print('El valor de a es: ', a, '     El valor de b es: ', b)
pendientes = []
ordenadas = []
func_error = []

step = 0
error_calc = 10

while error_calc> 0.02 and step <5000:
    grad = gradiente(x,y)
    step_a = grad[0]*learning_rate
    step_b = grad[1]*learning_rate
    a = a - step_a
    b = b - step_b
    error_calc = error(x,y)
    func_error.append(error_calc)
    step += 1
    pendientes.append(a)
    ordenadas.append(b)

print(f'El nuevo valor de a es: {a}')
print(f'El nuevo valor de b es: {b}')
print('El error es: ', error_calc)

y_nuevo = a*x + b
rectab = np.ones(len(ordenadas))
rectab *= b0
rectax = np.ones(len(pendientes))
rectax *= a0

fig, axs = plt.subplots(2,2, figsize=(10,10))
axs[0,0].plot(func_error)
axs[0,0].set_title('Magnitud de la función de error')
axs[0,0].set_xlabel('Iteraciones')
axs[0,0].set_ylabel('Magnitud del MSE')
axs[0,1].scatter(x,y)
axs[0,1].plot(x,y_nuevo, color='red')
axs[0,1].set_title('Datos originales y resultado del modelo')

axs[1,0].plot(pendientes)
axs[1,0].plot(rectax)
axs[1,0].set_title('Optimización de la pendiente')
axs[1,0].set_xlabel('Iteraciones')
axs[1,0].set_ylabel('Valor de la pendiente')
axs[1,1].plot(ordenadas)
axs[1,1].plot(rectab)
axs[1,1].set_title('Optimización de la ordenada')
axs[1,1].set_xlabel('Iteraciones')
axs[1,1].set_ylabel('Valor de la ordenada')

fig.tight_layout()
```
